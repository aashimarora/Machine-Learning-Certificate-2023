```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import toeplitz
from scipy.sparse.linalg import lsqr # import the least squares module
```

```python
sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)

noise_level = 0.25 # a number between 0 and 1

sig_noise = sig + noise_level * np.random.randn(len(sig))

sq_wave = np.ones(128)

corr = signal.correlate(sig_noise, sq_wave, mode='same') / 128
```

```python
plt.plot(sig,label='original',color='red')
```

```python
#plot the signal, the noisy signal and the correlated signal. 
plt.plot(sig,label='original',color='red')
plt.plot(sig_noise,label='noisy',color='green')
plt.plot(corr,label='correlated',color='blue')
plt.legend()
plt.show()
```
