```python
import numpy as np
import pandas as pd
from scipy.stats import bernoulli
```

## Missing Data Management - Probability Question
Construct a database with $100,000$ rows where each record has $100$ fields. Assume further that for each record, each of the $100$ fields has a $1$ per cent chance of being empty, i.e., its value is missing.

Remove all records with **two** or more empty fields and report the fraction of records that are removed.

```python
num_rows = 100000
data = np.ones(num_rows * 100)
data = pd.DataFrame(data.reshape(num_rows,100))
print(data)
```

```python
missing = bernoulli.rvs(p=0.01,size=num_rows * 100)
missing = missing.reshape(num_rows,100)
missing
```

```python
missing_rows, missing_cols = np.where(missing==1)
missing_rows,missing_cols
```

```python
data.values[missing_rows,missing_cols]=np.nan
data
```

```python
missing_values = data.isnull().sum(axis=1)
missing_values
```

```python
answer = np.sum(missing_values >= 2)/num_rows
```

```python
answer
```

The fraction of records that are removed = rows missing more than 2 values/ total rows = 26.64%
