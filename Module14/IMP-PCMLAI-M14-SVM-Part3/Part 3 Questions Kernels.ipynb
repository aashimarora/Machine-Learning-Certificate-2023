```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
import math
import warnings
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# use the following to plot SVMs
from mlxtend.plotting import plot_decision_regions # on terminal, install: pip.install mlxtend
```

#### Task 1: Support Vector Machines-based classifier on a non-separable dataset
**Load the 2-D data `Case3rings/X.npy` that has 150 rows and 2 columns, and the corresponding target `Case3rings/y.np`**

```python
X = np.load('Case3rings/X.npy')
y = np.load('Case3rings/Y.npy')
```

**Plot the two-dimensional points and color them so that points with the same class have the same color**

```python
n = 100
color = np.concatenate((np.repeat("blue", n), np.repeat("red",n/2)), axis=0)
plt.scatter(X[:,0], X[:,1], c = color, alpha= .5)
plt.xlabel(...)
plt.ylabel(...)
plt.title(...)
plt.show()
```

**Does a linear SVM based classifier (SVC) sound like a good idea? First discuss, then demonstrate by training, showing the confusion matrix, and plotting the boundary.**

- Answer: No, the data is not linearly separable. 

```python
clf = svm.SVC(kernel = 'linear')
clf.fit(X, y);
cm = confusion_matrix(y, clf.predict(X))
```

```python
import warnings
with warnings.catch_warnings(): #otherwise the package will complain there is no "boundary" as we classify all as '1'
    warnings.simplefilter("ignore")
    plot_decision_regions(X, y, clf=clf, legend=2, colors = "red,blue", markers= "o");
    ax=plt.gca();
    plt.show();
```

**In the linear SVC case above, why did our model end up classifying every instance with the very same prediction?**

- Answer: the data is not linearly separable, it took the majority vote?

```python
#Check the following: 
pd.value_counts(y)
```

**Try radial Kernel instead (with default parameters) and report the performance, and then plot the decision boundary**

Notes: 
- to change the kernel, try, e.g., `kernel = 'rbf'`
- to change the cost, try, e.g., `C = 10` 
- to change the scale parameter of RBF, try, e.g., `gamma = 4`
- see more options in the [official documentation](https://scikit-learn.org/stable/modules/svm.html)

```python
clf = svm.SVC(kernel='rbf')
clf.fit(X,y)
cm = confusion_matrix(y, clf.predict(X))
```

```python
plot_decision_regions(X, y, clf=clf, legend=2, colors = "red,blue", markers= "o");
```

**Try raidal Kernel (or RBF) with `gamma = 10`. Interpret the decision boundary.**

```python
clf = svm.SVC(kernel='rbf',gamma=10)
clf.fit(X,y)
cm = confusion_matrix(y, clf.predict(X))
```

```python
plot_decision_regions(X, y, clf=clf, legend=2, colors = "red,blue", markers= "o");
```

- Answer: gamma is too high so the model is overfitting

**Apply the same steps for a polynomial Kernel with degree 3**

```python
clf3 = svm.SVC(kernel='poly',degree=3)
clf3.fit(X,y)
cm = confusion_matrix(y, clf3.predict(X))
```

```python
plot_decision_regions(X, y, clf=clf3, legend=2, colors = "red,blue", markers= "o");
```

**Now try polynomial Kernel with degree 8 and cost 1**

```python
clf3 = svm.SVC(kernel='poly',degree=8,C=1)
clf3.fit(X,y)
cm = confusion_matrix(y, clf3.predict(X))
```

```python
plot_decision_regions(X, y, clf=clf3, legend=2, colors = "red,blue", markers= "o");
```

**Change the cost in the above case to 10**

```python
clf3 = svm.SVC(kernel='poly',degree=8,C=10)
clf3.fit(X,y)
cm = confusion_matrix(y, clf3.predict(X))
```

```python
plot_decision_regions(X, y, clf=clf3, legend=2, colors = "red,blue", markers= "o");
```

##### Notes:
- As you may have realized, the selection for a Kernel is not straightforward
- The best performing Kernels in this example are polynomial with degree 8 and cost 10, or RBF with gamma 10
- Moreover, even if we correctly guess RBf with gamma 10 but use a different cost, then our results may get significantly worse. 

**To illustrate them final point, run a radial Kernel SVM with gamma 10, but use a cost 0.3.**

```python
clf3 = svm.SVC(kernel='rbf',gamma=10,C=0.3)
clf3.fit(X,y)
cm = confusion_matrix(y, clf3.predict(X))
```

```python
plot_decision_regions(X, y, clf=clf3, legend=2, colors = "red,blue", markers= "o");
```
